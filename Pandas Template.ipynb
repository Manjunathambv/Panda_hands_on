```python
#3
## Start the code here
def summary_mean(x):
    x = pd.Series(x)
    return x.mean()
numeric_attributes = df.dtypes.loc[(df.dtypes == "int64") | (df.dtypes =="float64"),].index.tolist()
row_wise_mean = df[numeric_attributes].apply(summary_mean, axis = 1)
print("rown wise mean\n", row_wise_mean)
print("********************")
col_wise_mean = df[numeric_attributes].apply(summary_mean, axis = 0)
print("column wise mean\n", col_wise_mean)
# df[numeric_attributes]
## End here
```
